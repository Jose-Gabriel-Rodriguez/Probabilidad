```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
    numero_aleatorio = round( random.uniform(50, 100), 2 )
    fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
    numero_aleatorio2 = round( random.uniform(40, 93), 2 )
    fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

##  Maximo, mínimo, y rango  <a id='id4'> </a>

```python
maximo1 = max(fabrica1)
minimo1 = min(fabrica1)
rango1 = maximo1 - minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print( "Estadistica descriptiva de la Fabrica 1 \n", "Media =", media1,
     "\n Mediana =", mediana1, "\n Moda =", moda1, "\n Maximo = ", maximo1,
     "\n Minimo =", minimo1, "\n Rango =", rango1)
```

```python
maximo2 = max(fabrica2)
minimo2 = min(fabrica2)
rango2 = maximo1 - minimo2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print( "Estadistica descriptiva de la Fabrica 2 \n", "Media =", media2,
     "\n Mediana =", mediana2, "\n Moda =", moda2, "\n Maximo = ", maximo2,
     "\n Minimo =", minimo2, "\n Rango =", rango2)
```

## Dispersión de los datos   <a id='id5'> </a>

```python
desviacion_estandar_fabrica1 = stdev(fabrica1)
print(desviacion_estandar_fabrica1)
```

```python
desviacion_estandar_fabrica2 = stdev(fabrica2)
print(desviacion_estandar_fabrica2)
```

```python
plt.figure(figsize = (8,6) )
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media1 - desviacion_estandar_fabrica1, color='black', linestyle='dashed',
            linewidth=2, label=f'Desv. Estándar ={desviacion_estandar_fabrica1:.2f}')
plt.axvline(media1 + desviacion_estandar_fabrica1, color='black', linestyle='dashed',
            linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 1")
plt.legend()
plt.show()
```

# **Practica 3: Tablas de frecuencias**  <a id="id7"> </a>


```python
# basado en el LIBRO estadistica para administracion pagina 12
# creamos una lista vacia  que representa la edad de 100 personas
import random
edades = []
# Hacemos un ciclo para generar 100  numeros aleatorios qe representan la edad de las personas
for i in range(100):
  # generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  # numero_aleatorio
  numero_aleatorio = random.randint( 18, 70)
  # Agregamos el numero aleatorio a la lista de valores
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
# Crear un DataFrame de pandas con los números aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
# Calcular la tabla de distribución de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

# Imprimir la tabla de distribución de frecuencias
print("Tabla de Distribución de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

## Regla de sturges  <a id="id8"> </a>

1. Es una regla que sirve para calcular el número de clases o intervalos idóneo en los que se debe dividir un conjunto de datos.
2. Establece que **el número de clases es igual a uno más el logaritmo en base dos del número total de datos**.

$
c = 1 + log _2 (N)
$

donde $N$  es el tamaño de la muestra.

```python
# Importamos la librería math de funciones matemáticas
import math
```

```python
# Volvemos a usar la lista edades
print(edades)
```

```python
# Crear un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
# Calcular el número de intervalos utilizando la regla de Sturges
n = len(edades)
K = int( 1 + math.log2(n))
print("Valor de n = ", n , " \nNúmero de intervalos =", K)
```

```python
# Crear la tabla de frecuencia

tabla_frecuencia = pd.cut( df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# Generar columnas siguientes
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```

```python
plt.hist(edades, alpha = 0.5)
plt.title("Histograma de edades basado en la funcion de matplotlib con valores estandar")
```

```python
# Crear un histograma basado en la tabla de frecuencia
plt.hist(df['Edad'], bins=K, alpha = 0.5)
plt.title('Histograma de edades basado en la regla de Sturges')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```

### 9. Diagrama de Cajas y bigotes <a id="id9"> </a>

1. Las líneas se usan para indicar la variabilidad fuera de los cuartiles superior e inferior.
2. Los valores atípicos se representan a veces como puntos individuales que están en línea con los bigotes.
3.  **Si los datos son simétricos, la mediana estará en el centro de la caja**.
4. Si los datos están **sesgados**, la mediana estará más cerca de la parte superior o inferior de la caja.
5. Los extremos de arriba y abajo de la caja indican los **cuantiles, o percentiles, 25 y 75**. Estos dos cuantiles también se conocen como cuartiles, porque separan cuartos (25 %) de los datos. La longitud de la caja es la diferencia entre estos dos percentiles y se conoce como rango intercuartílico (IQR)
6. Los diagramas caja-bigotes son muy útiles para comparar dos o más conjuntos de datos.
