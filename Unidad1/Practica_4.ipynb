### 9. Diagrama de Cajas y bigotes <a id="id9"> </a>

1. Las líneas se usan para indicar la variabilidad fuera de los cuartiles superior e inferior.
2. Los valores atípicos se representan a veces como puntos individuales que están en línea con los bigotes.
3.  **Si los datos son simétricos, la mediana estará en el centro de la caja**.
4. Si los datos están **sesgados**, la mediana estará más cerca de la parte superior o inferior de la caja.
5. Los extremos de arriba y abajo de la caja indican los **cuantiles, o percentiles, 25 y 75**. Estos dos cuantiles también se conocen como cuartiles, porque separan cuartos (25 %) de los datos. La longitud de la caja es la diferencia entre estos dos percentiles y se conoce como rango intercuartílico (IQR)
6. Los diagramas caja-bigotes son muy útiles para comparar dos o más conjuntos de datos.

### Comparamos la producción de la fabrica 1 y 2 en un subplot

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

bplot1 = ax1.boxplot(fabrica1,
                     vert=True,  # vertical box alignment
                     patch_artist=True)  # fill with color
ax1.set_title('Producción fabrica 1')

bplot2 = ax2.boxplot(fabrica2,
                     vert=True,  # vertical box alignment
                     patch_artist=True)  # fill with color

ax2.set_title('Produccion fabrica 2')

plt.show()
```
